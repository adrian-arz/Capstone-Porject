# Introduction to Capstone Project. Problem description and Data Source

## 1. Problem description

There are many religions that do not allow every kind of food or beverages, as jewish or islam. It is something minor when in your residence city, but it can became sucha a problem when travelling. For this exercise, my aim is to help jewish people when doing citysightseeing in New York, finding the best kosher restaurants at each aerea and looking for the best areas to host. Same exercise could be done for other cities just changing the neighbourhood data, that can be easily taken from wikipedia.
